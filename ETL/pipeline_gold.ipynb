```python
# Carregar arquivos
clientes_df = pd.read_parquet(f'{caminho_silver}/clientes.parquet')
vendedores_df = pd.read_parquet(f'{caminho_silver}/vendedores.parquet')
itens_pedidos_df = pd.read_parquet(f'{caminho_silver}/itens_pedidos.parquet')
geolocalizacao_df = pd.read_parquet(f'{caminho_silver}/geolocalizacao.parquet')
pagamentos_df = pd.read_parquet(f'{caminho_silver}/pagamentos.parquet')
pedidos_df = pd.read_parquet(f'{caminho_silver}/pedidos.parquet')
produtos_df = pd.read_parquet(f'{caminho_silver}/produtos.parquet')
```

# II - Preparação e Enriquecimento dos Dados



#### 2.1 - Cálculo do Volume e Densidade dos Produtos


```python
# Calcular o volume dos produtos em cm³
produtos_df['product_volume_cm3'] = (
    produtos_df['product_length_cm'] *
    produtos_df['product_height_cm'] *
    produtos_df['product_width_cm']
)

# Calcular a densidade dos produtos em g/cm³
produtos_df['product_density_g_per_cm3'] = (
    produtos_df['product_weight_g'] / produtos_df['product_volume_cm3']
)
```

### 2.2 - Mesclando Características dos Produtos nos Itens de Pedido


```python

# Adicionar informações de produto aos itens pedidos
itens_pedidos_df = itens_pedidos_df.merge(
    produtos_df[['product_id', 'product_category', 'product_volume_cm3', 'product_density_g_per_cm3']],
    on='product_id',
    how='left'
)
```

### 2.3 - Adicionando Informações de Clientes aos Pedidos


```python

# Adicionar informações de cliente aos pedidos
pedidos_df = pedidos_df.merge(
    clientes_df[['customer_id', 'customer_unique_id']],
    on='customer_id',
    how='left'
)
```

### 2.4 - Adicionando 'customer_unique_id' aos Itens de Pedido


```python
# Adiciona 'customer_unique_id' ao 'itens_pedidos_df' mesclando com 'pedidos_df'
itens_pedidos_df = itens_pedidos_df.merge(
    pedidos_df[['order_id', 'customer_id', 'customer_unique_id']],
    on='order_id',
    how='left'
)
```

### 2.5 - Calculando a Margem Bruta



```python
# Calcular a margem bruta de cada item pedido
itens_pedidos_df['margem_bruta'] = itens_pedidos_df['price'] - itens_pedidos_df['freight_value']
```

### 2.6 - Criando a Tabela de Vendas


```python
# Criar a tabela de vendas (tabela fato)
vendas_df = itens_pedidos_df[[
    'order_id',
    'order_item_id',
    'product_id',
    'seller_id',
    'customer_id',
    'customer_unique_id',
    'price',
    'freight_value',
    'margem_bruta',
    'shipping_limit_date'
]]
```

### 2.7 - Agrupando prefixos dos CEP's

```python
# Agregar a tabela de geolocalização por prefixo de CEP
geolocalizacao_agg = geolocalizacao_df.groupby('geolocation_zip_code_prefix').agg({
    'geolocation_lat': 'mean',
    'geolocation_lng': 'mean'
}).reset_index()
```

# III - Criação das Tabelas de Dimensões


### 3.1 - Dimensão Produto


```python
# Criar a tabela de dimensão produto
dim_produto_df = produtos_df[[
    'product_id',
    'product_category',
    'product_name_length',
    'product_description_length',
    'product_photos_qty',
    'product_weight_g',
    'product_length_cm',
    'product_height_cm',
    'product_width_cm',
    'product_volume_cm3',
    'product_density_g_per_cm3'
]].drop_duplicates()

# Renomear colunas para padronização
dim_produto_df.rename(columns={'product_id': 'id_produto'}, inplace=True)
```

### 3.2 - Dimensão Cliente



```python
# Mesclar clientes com a tabela de geolocalização agregada
dim_cliente_df = clientes_df.merge(
    geolocalizacao_agg,
    left_on='customer_zip_code_prefix',
    right_on='geolocation_zip_code_prefix',
    how='left'
)

# Selecionar colunas relevantes e remover duplicatas
dim_cliente_df = dim_cliente_df[[
    'customer_unique_id',
    'customer_zip_code_prefix',
    'customer_city',
    'customer_state',
    'geolocation_lat',
    'geolocation_lng'
]].drop_duplicates()

# Renomear colunas para padronização
dim_cliente_df.rename(columns={
    'customer_unique_id': 'id_cliente',
    'customer_zip_code_prefix': 'cep_cliente',
    'customer_city': 'cidade_cliente',
    'customer_state': 'estado_cliente',
    'geolocation_lat': 'latitude',
    'geolocation_lng': 'longitude'
}, inplace=True)
```

### 3.3 - Dimensão Vendedor


```python
# Mesclar geolocalização com vendedores usando o zip code
dim_vendedor_df = vendedores_df.merge(
    geolocalizacao_df,
    left_on='seller_zip_code_prefix',
    right_on='geolocation_zip_code_prefix',
    how='left'
)

# Selecionar colunas relevantes e remover duplicatas
dim_vendedor_df = dim_vendedor_df[[
    'seller_id',
    'seller_zip_code_prefix',
    'seller_city',
    'seller_state',
    'geolocation_lat',
    'geolocation_lng'
]].drop_duplicates()

# Renomear colunas para padronização
dim_vendedor_df.rename(columns={
    'seller_id': 'id_vendedor',
    'seller_zip_code_prefix': 'cep_vendedor',
    'seller_city': 'cidade_vendedor',
    'seller_state': 'estado_vendedor',
    'geolocation_lat': 'latitude',
    'geolocation_lng': 'longitude'
}, inplace=True)
```

### 3.4 - Dimensão Tempo


```python

# Definir o período para a dimensão tempo
min_data = pedidos_df['order_purchase_timestamp'].min().date()
max_data = pedidos_df['order_purchase_timestamp'].max().date()

# Criar a tabela de dimensão tempo
dim_tempo_df = pd.DataFrame({'data': pd.date_range(start=min_data, end=max_data)})

# Adicionar atributos de tempo
dim_tempo_df['id_tempo'] = dim_tempo_df['data']
dim_tempo_df['ano'] = dim_tempo_df['data'].dt.year
dim_tempo_df['mes'] = dim_tempo_df['data'].dt.month
dim_tempo_df['dia'] = dim_tempo_df['data'].dt.day
dim_tempo_df['nome_mes'] = dim_tempo_df['data'].dt.strftime('%B')
dim_tempo_df['trimestre'] = dim_tempo_df['data'].dt.quarter
dim_tempo_df['dia_semana'] = dim_tempo_df['data'].dt.weekday + 1
dim_tempo_df['nome_dia_semana'] = dim_tempo_df['data'].dt.strftime('%A')
```

### 3.5 - Dimensão Pagamento



```python
# Criar a tabela de dimensão pagamento
dim_pagamento_df = pagamentos_df[['payment_type']].drop_duplicates()

# Adicionar um identificador para cada tipo de pagamento
dim_pagamento_df['id_pagamento'] = dim_pagamento_df.reset_index().index + 1

# Renomear colunas para padronização
dim_pagamento_df.rename(columns={'payment_type': 'tipo_pagamento'}, inplace=True)
```

# IV - Ajustes Finais na Tabela Vendas


### 4.1 - Adicionando Data de Compra e Chaves de Tempos

```python
# Adicionar data de compra à tabela de vendas
vendas_df = vendas_df.merge(
    pedidos_df[['order_id', 'order_purchase_timestamp']],
    on='order_id',
    how='left'
)
vendas_df['data_compra'] = vendas_df['order_purchase_timestamp'].dt.date

# Remover colunas desnecessárias
vendas_df.drop(columns=['order_purchase_timestamp'], inplace=True)

# Adicionar chave de tempo à tabela de vendas
vendas_df['id_tempo'] = vendas_df['data_compra']
```

### 4.2 - Mesclando Pagamentos para Obter o Tipo de Pagamento


```python
# Adicionar informações de pagamento à tabela de vendas
vendas_df = vendas_df.merge(
    pagamentos_df[['order_id', 'payment_type']],
    on='order_id',
    how='left'
)

# Mesclar com a dimensão pagamento para obter o id_pagamento
vendas_df = vendas_df.merge(
    dim_pagamento_df,
    left_on='payment_type',
    right_on='tipo_pagamento',
    how='left'
)
```

### 4.3 - Removendo Colunas Desnecessárias


```python
# Remover colunas intermediárias não mais necessárias
vendas_df.drop(columns=['data_compra', 'payment_type', 'tipo_pagamento'], inplace=True)
```

### 4.4 - Tradução colunas

```python
# Listar os nomes das colunas de cada DataFrame
colunas_dim_produto = dim_produto_df.columns.tolist()
colunas_dim_cliente = dim_cliente_df.columns.tolist()
colunas_dim_vendedor = dim_vendedor_df.columns.tolist()
colunas_dim_tempo = dim_tempo_df.columns.tolist()
colunas_dim_pagamento = dim_pagamento_df.columns.tolist()

# Exibir os nomes das colunas
{
    "dim_produto": colunas_dim_produto,
    "dim_cliente": colunas_dim_cliente,
    "dim_vendedor": colunas_dim_vendedor,
    "dim_tempo": colunas_dim_tempo,
    "dim_pagamento": colunas_dim_pagamento,
}
```

```python
# Renomear as colunas da tabela dim_produto_df

dim_produto_df.rename(columns={
    'product_category': 'categoria_produto',
    'product_name_length': 'tamanho_nome_produto',
    'product_description_length': 'tamanho_descricao_produto',
    'product_photos_qty': 'quantidade_fotos_produto',
    'product_weight_g': 'peso_produto_g',
    'product_length_cm': 'comprimento_produto_cm',
    'product_height_cm': 'altura_produto_cm',
    'product_width_cm': 'largura_produto_cm',
    'product_volume_cm3': 'volume_produto_cm3',
    'product_density_g_per_cm3': 'densidade_produto_g_por_cm3'
}, inplace=True)



# Renomear as colunas da tabela vendas_df
vendas_df.rename(columns={
    'order_id': 'id_pedido',
    'order_item_id': 'id_item_pedido',
    'product_id': 'id_produto',
    'seller_id': 'id_vendedor',
    'customer_id': 'id_cliente',
    'customer_unique_id': 'id_cliente_unico',
    'price': 'preco',
    'freight_value': 'valor_frete',
    'margem_bruta': 'margem_bruta',
    'shipping_limit_date': 'data_limite_envio',
    'id_tempo': 'id_tempo',
    'id_pagamento': 'id_pagamento'
}, inplace=True)
```
