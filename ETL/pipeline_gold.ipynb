# II - Preparação e Enriquecimento dos Dados



#### 2.1 - Cálculo do Volume e Densidade dos Produtos


```python
produtos_df['product_volume_cm3'] = (
    produtos_df['product_length_cm'] *
    produtos_df['product_height_cm'] *
    produtos_df['product_width_cm']
)
produtos_df['product_density_g_per_cm3'] = (
    produtos_df['product_weight_g'] / produtos_df['product_volume_cm3']
)
```

### 2.2 - Mesclando Características dos Produtos nos Itens de Pedido


```python
itens_pedidos_df = itens_pedidos_df.merge(
    produtos_df[['product_id', 'product_category', 'product_volume_cm3', 'product_density_g_per_cm3']],
    on='product_id',
    how='left'
)
```

```python

# Realizando o join para adicionar a coluna 'customer_id' de pedidos_df a itens_pedidos_df
itens_pedidos_df = itens_pedidos_df.merge(
    pedidos_df[['order_id', 'customer_id']],
    on='order_id',
    how='left'
)

print("Colunas em itens_pedidos_df:", itens_pedidos_df.columns)
print("Colunas em pedidos_df:", pedidos_df.columns)
```

### 2.3 - Calculando o Tempo de Entrega e Atraso nos Pedidos


```python
pedidos_df['tempo_entrega_dias'] = (
    pedidos_df['order_delivered_customer_date'] - pedidos_df['order_purchase_timestamp']
).dt.days
pedidos_df['atraso_entrega_dias'] = (
    pedidos_df['order_delivered_customer_date'] - pedidos_df['order_estimated_delivery_date']
).dt.days
```

### 2.4 - Adicionando Informações de Clientes aos Pedidos

```python
pedidos_df = pedidos_df.merge(
    clientes_df[['customer_id', 'customer_unique_id', 'customer_state']],
    on='customer_id',
    how='left'
)
```

### 2.5 - Consolidando Pagamentos por Pedido


```python
pagamentos_agg_df = pagamentos_df.groupby('order_id').agg(
    valor_total_pagamento=('payment_value', 'sum'),
    tipo_pagamento=('payment_type', lambda x: x.mode()[0]),
    parcelas=('payment_installments', 'max')
).reset_index()
```

# III - Cálculo de KPI's

## 3.1 - KPIs de Desempenho Financeiro e Receita

```python
itens_pedidos_df.columns
```

```python
## Receita Bruta Total
receita_bruta_total = itens_pedidos_df['price'].sum()

## Receita Bruta por Cliente
itens_pedidos_df = itens_pedidos_df.merge(
    pedidos_df[['order_id', 'customer_unique_id']], on='order_id', how='left'
)
receita_bruta_cliente_df = itens_pedidos_df.groupby('customer_unique_id').agg(
    receita_bruta=('price', 'sum')
).reset_index()

## Receita por Produto/Categoria
receita_categoria_df = itens_pedidos_df.groupby('product_category').agg(
    receita=('price', 'sum')
).reset_index()

## Ticket Médio por Pedido
ticket_medio = itens_pedidos_df.groupby('order_id')['price'].sum().mean()

## Margem Bruta (Receita Bruta - Custo de Frete)
itens_pedidos_df['margem_bruta'] = itens_pedidos_df['price'] - itens_pedidos_df['freight_value']
margem_bruta_total = itens_pedidos_df['margem_bruta'].sum()

## Contribuição por Categoria de Produto
contribuicao_categoria_df = receita_categoria_df.copy()
contribuicao_categoria_df['percentual_contribuicao'] = (
    contribuicao_categoria_df['receita'] / receita_bruta_total * 100
)
```

```python
receita_bruta_cliente_df.columns
```

## 3.2 - KPIs de Eficiência de Logística e Frete


```python
## Custo Total e Médio de Frete
custo_total_frete = itens_pedidos_df['freight_value'].sum()
custo_medio_frete = itens_pedidos_df['freight_value'].mean()

## Frete Médio por Região
frete_regiao_df = itens_pedidos_df.merge(
    pedidos_df[['order_id', 'customer_state']], on='order_id', how='left'
).groupby('customer_state').agg(
    frete_medio=('freight_value', 'mean')
).reset_index()

## Taxa de Frete Grátis
taxa_frete_gratis = (
    (itens_pedidos_df[itens_pedidos_df['freight_value'] == 0].shape[0]) / itens_pedidos_df.shape[0]
) * 100

## Tempo Médio de Entrega
tempo_medio_entrega = pedidos_df['tempo_entrega_dias'].mean()

## Taxa de Atraso nas Entregas e Dias de Atraso Médio
pedidos_df['entrega_atrasada'] = pedidos_df['atraso_entrega_dias'] > 0
taxa_atraso_entrega = pedidos_df['entrega_atrasada'].mean() * 100
dias_atraso_medio = pedidos_df[pedidos_df['entrega_atrasada']]['atraso_entrega_dias'].mean()

## Taxa de Entregas Antecipadas
pedidos_df['entrega_antecipada'] = pedidos_df['atraso_entrega_dias'] < 0
taxa_entrega_antecipada = pedidos_df['entrega_antecipada'].mean() * 100

```

## 3.3 - KPIs de Comportamento do Cliente e Pagamentos


```python
## Frequência de Compra por Cliente
frequencia_compra_df = pedidos_df.groupby('customer_unique_id').agg(
    frequencia_compra=('order_id', 'count')
).reset_index()

## Valor de Vida do Cliente (CLV)
clv_df = receita_bruta_cliente_df.merge(
    frequencia_compra_df, on='customer_unique_id', how='left'
)
clv_df['clv'] = clv_df['receita_bruta'] * clv_df['frequencia_compra']

## Preferência de Método de Pagamento
preferencia_pagamento_df = pagamentos_df['payment_type'].value_counts(normalize=True).reset_index()
preferencia_pagamento_df.columns = ['payment_type', 'percentual']

## Percentual de Parcelamentos
percentual_parcelamentos = (
    pagamentos_df[pagamentos_df['payment_installments'] > 1].shape[0] / pagamentos_df.shape[0]
) * 100
```

## 3.4 - KPIs de Produtos e Logística de Armazenagem


```python
## Volume e Densidade Médios por Categoria de Produto
volume_densidade_categoria_df = produtos_df.groupby('product_category').agg(
    volume_medio=('product_volume_cm3', 'mean'),
    densidade_media=('product_density_g_per_cm3', 'mean')
).reset_index()

## Top 10 Produtos por Receita
receita_produto_df = itens_pedidos_df.groupby('product_id').agg(
    receita=('price', 'sum')
).reset_index()
top_10_produtos_df = receita_produto_df.merge(
    produtos_df[['product_id', 'product_category']], on='product_id', how='left'
).sort_values(by='receita', ascending=False).head(10)
```

# IV - Salvando na Camada Gold
